```python
import os
from dotenv import load_dotenv
import pathlib
import duckdb

load_dotenv()

TEST_HAND_FILE = os.getenv('TEST_HAND_FILE')
DUCKDB_FILE = os.getenv('DUCKDB_FILE')
PARQUET_DATA_DIR = pathlib.Path(os.getenv('PARQUET_DATA_DIR'))

HANDS_PARQUET_PATH = PARQUET_DATA_DIR / 'Hands' / '*.parquet'
PLAYERS_PARQUET_PATH = PARQUET_DATA_DIR / 'Players' / '*.parquet'
ACTIONS_PARQUET_PATH = PARQUET_DATA_DIR / 'Actions' / '*.parquet'
```

```python
con = duckdb.connect(DUCKDB_FILE)

con.execute(f"""
    CREATE TABLE IF NOT EXISTS brnz_actions AS
    SELECT * FROM '{ACTIONS_PARQUET_PATH}'
""")

con.execute(f"""
    CREATE TABLE IF NOT EXISTS brnz_hands AS
    SELECT * FROM '{HANDS_PARQUET_PATH}'
""")

con.execute(f"""
    CREATE TABLE IF NOT EXISTS brnz_players AS
    SELECT * FROM '{PLAYERS_PARQUET_PATH}'
""")
```
