```python
import os
from dotenv import load_dotenv
import pathlib
import duckdb

load_dotenv()

TEST_HAND_FILE = os.getenv('TEST_HAND_FILE')
DUCKDB_FILE = os.getenv('DUCKDB_FILE')
PARQUET_DATA_DIR = pathlib.Path(os.getenv('PARQUET_DATA_DIR'))

HANDS_PARQUET_PATH = PARQUET_DATA_DIR / 'Hands' / '*.parquet'
PLAYERS_PARQUET_PATH = PARQUET_DATA_DIR / 'Players' / '*.parquet'
ACTIONS_PARQUET_PATH = PARQUET_DATA_DIR / 'Actions' / '*.parquet'
```

```python
def duckdb_run_query(query: str):
    con = duckdb.connect(DUCKDB_FILE)
    result = con.execute(query).fetchdf()
    con.close()
    return result
```

```python
# Check the player ID to see if they have more hands than a single file. This will show if the player hashes are consistent across files.
df_player_hash_check = duckdb_run_query(f"""
    SELECT player_name, COUNT(DISTINCT hand_id) AS num_hands
    FROM slv_players
    GROUP BY player_name
    HAVING num_hands > 1
    ORDER BY num_hands DESC
""")

display(df_player_hash_check)
```

```python
df_hands = duckdb_run_query(f"""
    SELECT *
    FROM slv_hands
    LIMIT 500
""")

display(df_hands)
```
